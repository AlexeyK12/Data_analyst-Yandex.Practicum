В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Очень много пропусков. Высота потолка в домах обычно варьируется от 2,2 м до 8 м (в пентхаусах со вторым светом). Из описания видим, что присутствуют квартиры с высотой потолка от 1 до 100 м.

```python
data['ceiling_height'].hist(bins=50, figsize=(15,5), range=(2.2,10))
plt.title('Высота потолка')
```

Есть предположение, что в значениях чрезмерно "высоких" квартир, неверно стоит запятая. Посмотрим на параметры квартир, высота потолка в которых больше 22 м и меньше 80 м.

```python
display(data.query('ceiling_height > 22 and ceiling_height < 80'))
```

Предположение выглядит правдоподобно. Исправим значения, для этого разделим их на 10.

```python
index_ceiling_height = data.query('ceiling_height > 22 and ceiling_height < 80').index
data.loc[index_ceiling_height, 'ceiling_height'] = data.loc[index_ceiling_height, 'ceiling_height'] / 10
```

Удалим строки с аномальными значениями( меньше 2,2 м и выше 8 м)

```python
index_list = data.query('ceiling_height > 80 or ceiling_height < 2.2 or 8 < ceiling_height < 22').index
data.drop(index_list, inplace=True, axis=0)
data.reset_index(inplace=True, drop=True)
print(data['ceiling_height'].describe())
```

```python
print('Количество пропусков:',data['ceiling_height'].isna().sum())
```

Поскольку высота потолка сильно отличается, заменим NaN на медиану = 2,65 м.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(2.65)
print("Кол-во пропусков =", data['ceiling_height'].isna().sum())
data['ceiling_height'].describe()
```

## Расчёты и добавление результатов в таблицу

Посчитаем и добавим в таблицу:
- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры: варианты - первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.

**Цена квадратного метра**

```python
data['sqare_metr_price'] = data['last_price']/data['total_area']
data['sqare_metr_price'] = data['sqare_metr_price'].astype(int)
data.info()
data.head()
```

**День недели, месяц и год публикации объявления**

```python
data['day_exposition'] = data['first_day_exposition'].dt.weekday
data['month_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year
display(data[['day_exposition', 'month_exposition', 'year_exposition']].head(10))
data[['day_exposition', 'month_exposition', 'year_exposition']].describe()
data.info()
data.head()
```

**Категории этажей**

```python
def floor_category(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
data['floor_category'] = data.apply(floor_category, axis=1)    
display(data.head())
data.info()
```

**Соотношение жилой и общей площади**

```python
data['living_total_area'] = round(data['kitchen_area'] / data['total_area'],2)
data['kitchen_total_area'] = round(data['kitchen_area'] / data['total_area'],2)
data.head()
```

```python
for col in ['total_area', 'last_price', 'rooms', 'ceiling_height']:
    spb_center.plot(y=col, kind='hist', grid=True, figsize=(14,3), bins=75, alpha=0.8, legend=True, label='ЦЕНТР - '+col)
```

```python
def corr_spb(column):   
    plt.show()
    print('Корелляция цены в Центральной зоне от',column , spb_center['last_price'].corr(spb_center[column]))
    spb_center.plot(x=column, y='last_price', kind='scatter', grid=True, figsize=(14,4))
    plt.show()
    
for сol in ['rooms', 'floor', 'city_centers_nearest', 'days_exposition', 'month_exposition', 'year_exposition']:
    corr_spb(сol)


cat_fl_mean_spb_center = spb_center.groupby('floor_category')['last_price'].mean()

cat_fl_mean_spb_center.plot(x=cat_fl_mean.index, y=cat_fl_mean.values, kind='bar', grid=True, figsize=(5,4), alpha=0.5)
plt.show()
```
